```python
import pandas as pd
import os
```

```python
os.chdir("/cluster2/home/zeyu/Projects/Program/cross_species_integration/Matching/Methods/HL_O2O/pancreatic_islet_cells")
human_mice=pd.read_csv("/cluster2/home/zeyu/Projects/Program/cross_species_integration/Matching/Methods/HL_O2O/initial/HL_O2O_Results/human_mice_union_greedy.csv", header=0, dtype=str,usecols=[0,1])
mice_pig=pd.read_csv("/cluster2/home/zeyu/Projects/Program/cross_species_integration/Matching/Methods/HL_O2O/initial/HL_O2O_Results/mice_pig_union_greedy.csv", header=0, dtype=str,usecols=[0,1])
```

```python
print(human_mice.shape)
print(human_mice.iloc[0:5,:])
```

```python
print(mice_pig.shape)
print(mice_pig.iloc[0:5,:])
```

```python
human_mice = human_mice.dropna(subset=[human_mice.columns[0], human_mice.columns[1]])#human_mice mice_pig
mice_pig = mice_pig.dropna(subset=[mice_pig.columns[0], mice_pig.columns[1]])
print(human_mice.shape)
print(mice_pig.shape)
```

```python
print(human_mice.iloc[0:5,:])
print(mice_pig.iloc[0:5,:])
```

```python
human_mice=human_mice.rename(columns={human_mice.columns[0]:"human gene name"})
human_mice=human_mice.rename(columns={human_mice.columns[1]:"mice gene name"})
mice_pig=mice_pig.rename(columns={mice_pig.columns[0]:"mice gene name"})
mice_pig=mice_pig.rename(columns={mice_pig.columns[1]:"pig gene name"})
```

```python
print(human_mice.shape)
print(mice_pig.shape)
```

```python
human_mice = human_mice.drop_duplicates()
mice_pig=mice_pig.drop_duplicates()
```

```python
human_mice_pig = human_mice.merge(mice_pig, on="mice gene name", how="outer")
human_mice_pig = human_mice_pig.dropna()
print(human_mice_pig.shape)
print(human_mice_pig.iloc[0:5,:])
```

```python
human_mice_pig.to_csv("human_mice_pig.csv",index=False)
```
